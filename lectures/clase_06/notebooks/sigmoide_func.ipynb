```python
import numpy as np # algebra lineal + matematicas
import matplotlib.pyplot as plt # graficar 
```

```python
z = np.linspace(-100,100,2000)
```

```python
z[0]
```

```python
resultado = []
for i in range(len(z)):
    resultado.append(1/(1+np.exp(-z[i])))
```

```python
# declaracion del metodo
def sigmoide(z: np.array)->np.array:
    return 1/(1+np.exp(-z))
```

```python
# ejecutar el metodo
sigmoide(z)
```

```python
plt.plot(z,sigmoide(z))
plt.grid()
```
