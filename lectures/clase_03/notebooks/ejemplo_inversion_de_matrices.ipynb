```python
import numpy as np
```

```python
X = np.array([[1,1,1],[1,2,4],[1,3,9]])
```

```python
y = np.array([[1],[0.5],[2]])
```

```python
sp = np.linalg.inv(X.T@X)
```

```python
coeff = sp@X.T@y
```
