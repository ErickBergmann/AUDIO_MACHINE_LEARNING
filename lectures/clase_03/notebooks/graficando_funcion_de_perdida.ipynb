```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
```

Para nuestro ejemplo

```python
x = np.array([1,2,3])
y = np.array([1, 0.5, 2])
```

```python
theta_1 = np.linspace(-0.7,0.7,100)
theta_0 = np.linspace(-0.17,0.17, 100)
```

Creamos todas las posibles combinaciones de theta_0 y theta_1

```python
z = []
for t0 in theta_0:
  for t1 in theta_1:
    z.append(((1/len(x))*np.sum((t0 + t1*x) - y)**2,[t0,t1]))
```

```python
# Create figure outside the loop
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for i in range(0,len(z)):
  ax.scatter(z[i][1][0], z[i][1][1], z[i][0], c='r', marker='o')

  # Set labels for the axes
  ax.set_xlabel('X Label')
  ax.set_ylabel('Y Label')
  ax.set_zlabel('Z Label')

  # Change the viewing angle
  elevation_angle = 30  # Vertical viewing angle in degrees
  azimuth_angle = 45    # Horizontal viewing angle in degrees
  ax.view_init(elev=elevation_angle, azim=azimuth_angle)

# Show the plot
plt.show()
```
