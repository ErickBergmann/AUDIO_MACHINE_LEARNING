```python
import numpy as np
import typing
from typing import List, Tuple

def svd_nico(X: np.array ,full_matrices = False, more_matrices = True)-> Tuple[np.array]:
  '''
    SVD de mi forma
  '''
  U,S,V = np.linalg.svd(X,full_matrices = full_matrices)
  if not more_matrices:
    return S
  if more_matrices:
    return U,S,V
```

```python
svd_nico(M, more_matrices = True)
```

```python
def fib(n): 
    """Imprime n términos de la serie Fibonacci."""
    a, b = 0, 1
    while a < n:
        print(a, end=' ')
        a, b = b, a+b
    print()
```

```python
fib(4)
```

Ejemplo 3

El primer argumento de esta función es proporcionado por el usuario.
Los otros dos tienen valores por defecto, que pueden ser modificados al darle los valores de entrada a la función si asi se quiere.

```python
def ask_ok(prompt, retries=4, reminder='Intentalo de nuevo!'):
    while True:
        ok = input(prompt)
        if ok in ('s', 'si', 'sip'):
            return True
        if ok in ('n', 'no', 'nop', 'nope'):
            return False
        retries = retries - 1
        if retries < 0:
            raise ValueError('respuesta inválida')
        print(reminder)
```

```python
ask_ok('te quieres salir del curso?')
```

```python
ask_ok('te quieres salir del curso?')
```

```python
ask_ok('te quieres salir del curso?')
```

Argumentos posicionales

```python
def parrot(voltage, state='a stiff', action='voom', type='Norwegian Blue'):
    print("-- This parrot wouldn't", action, end=' ')
    print("if you put", voltage, "volts through it.")
    print("-- Lovely plumage, the", type)
    print("-- It's", state, "!")
```

```python
parrot(1000)                                          # 1 positional argument
```

```python
parrot(voltage=1000)                                  # 1 keyword argument
```

```python
parrot(voltage=1000000, action='VOOOOOM')             # 2 keyword arguments
```

```python
parrot('a million', 'bereft of life', 'jump')         # 3 positional arguments
```

```python
parrot('a thousand', state='pushing up the daisies')  # 1 positional, 1 keyword
```

Formas Inválidas

```python
parrot()                     # required argument missing
parrot(voltage=5.0, 'dead')  # non-keyword argument after a keyword argument
parrot(110, voltage=220)     # duplicate value for the same argument
parrot(actor='John Cleese')  # unknown keyword argument
```

Entregando argumentos a través de diccionarios

```python
def cheeseshop(kind, *arguments, **keywords):
    print("-- Do you have any", kind, "?")
    print("-- I'm sorry, we're all out of", kind)
    for arg in arguments:
        print(arg)
    print("-" * 40)
    for kw in keywords:
        print(kw, ":", keywords[kw])
```

```python
cheeseshop("Limburger", "It's very runny, sir.",
           "It's really very, VERY runny, sir.",
           shopkeeper="Michael Palin",
           client="John Cleese",
           sketch="Cheese Shop Sketch")
```

```python
def recapitulando(texto, *argumentos, **palabras_clave):
  print("Hemos aprendido algo sobre", texto, "?")
  print("Creo que no sé si hemos aprendido algo sobre", texto)
  for arg in argumentos:
    print(arg)
  print("-" * 40)
  for kw in palabras_clave:
    print(kw, ":", palabras_clave[kw])
```

```python
recapitulando("programación","increíble, todo este tiempo perdido",integrantes = ' Profesor', numero_de_estudiantes = 10)
```

# Funciones Lambda

```python
def sumador_incremental(n):
    return lambda x: x + n
```

```python
f = sumador_incremental(17)
```

```python
f(0)
```

```python
f(20)
```

```python
f(23)
```

```python
x = lambda a, b : a * b
print(x(5, 6))
```
