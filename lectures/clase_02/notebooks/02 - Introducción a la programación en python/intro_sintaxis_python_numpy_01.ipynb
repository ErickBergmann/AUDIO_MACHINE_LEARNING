**Operaciones de álgebra lineal** 

*Producto Interno*

Una operación básica es el producto punto entre dos vectores. Matemáticamente dados dos vectores $\mathbf{x}$ e $\mathbf{y}\in \mathbb{R}^{N}$ (dos vectores de dimensión $N$ o con $N$ elementos) el producto $f: \mathbb{R}^{N}\times \mathbb{R}^{N}\rightarrow \mathbb{R}$ es dado por:

\begin{equation}
  f(\mathbf{x},\mathbf{y}) = \langle \mathbf{x}, \mathbf{y} \rangle = \displaystyle \sum_{i = 1}^{N} x_{i}y_{i}
\end{equation}

que representa la multiplicación componente a componente de los elementos de los vectores $\mathbf{x}$ e $\mathbf{y}$ y luego la suma de estas multiplicaciones, por ejemplo:

\begin{equation}
  f([1,2],[-1,4]) = \langle [1,2], [-1,4]\rangle = \displaystyle \sum_{i = 1}^{N} x_{i}y_{i} = 1\cdot -1 + 2 \cdot 4 = -1 + 8 = 7
\end{equation}

comunmente se utiliza el símbolo $\langle \cdot, \cdot \rangle$

En $\mathtt{numpy}$ podemos calcularlo así:

````python
numpy.dot(x_vector,x_vector.T)
````

la función ```.T``` aplica la transposición del vector (Lo transforma en un vector columna). Esto porque, debido a las reglas del álgebra lineal, las dimensiones internas de los vectores deben ser iguales para poder aplicar esta operación, por ejemplo en este caso: $(1,\mathbf{4}) \times (\mathbf{4},1)$, los números en negrito se llaman *dimensiones internas* y ellas deben coincidir. Si no realizamos transposición tendríamos  $(1,\mathbf{4}) \times (\mathbf{1},4)$ lo que arrojaría **el siguiente error**:

````python
numpy.dot(x_vector,x_vector)
````

*Multiplicación de Matrices*

En este caso también necesitamos respetar la regla de las dimensiones internas!.
Matemáticamente dadas dos matrices $A\in \mathbb{R}^{K\times N}$ y $B\in \mathbb{R}^{N\times D}$, el producto $g: \mathbb{R}^{K\times N}\times \mathbb{R}^{N\times D}\rightarrow {R}^{K \times D}$ es dado por:

\begin{equation}
  g\left(A,B \right) = (A B)_{ij} = \displaystyle \sum_{k = 1}^{N} a_{ik}b_{kj}
\end{equation}

para "memorizar" esta fórmula, generalmente se piensa en "multiplicar filas de la matriz $A$ por las columnas de la matriz $B$" pero esta cuando hablamos de "multiplicar" nos referimos al producto interno. Por ejemplo:

````python
matriz_de_7 = 7*numpy.ones((6,4))
````

````python
matriz_de_7.shape
````
