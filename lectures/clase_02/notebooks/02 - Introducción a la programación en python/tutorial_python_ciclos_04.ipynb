# Ciclos $\mathtt{While}$

````python
# Serie de Fibonacci:
# La suma de dos elementos define al siguiente
a, b = 0, 1
while a < 10:
    print(a)
    a, b = b, a+b
````

````python
# Para evitar el salto de línea podemos utilizar el keyword ```end```
a, b = 0, 1
while a < 1000:
    print(a, end=',')
    a, b = b, a+b
````

````python
x = int(input("Por favor ingresa un número entero: "))
````

````python
if x < 0:
    x = 0
    print('Número negativo o cero')
elif x == 0:
    print('cero')
elif x == 1:
    print('uno')
else:
    print('diferente de 0,1 o negativos')
````

# Diccionarios o colecciones

````python
# Create a sample collection
users = {'Hans': 'active', 'Éléonore': 'inactive', '景太郎': 'active'}

# Strategy:  Iterate over a copy
for user, status in users.copy().items():
    if status == 'inactive':
        del users[user]
````

````python
users
````

````python
# Strategy:  Create a new collection
active_users = {}
for user, status in users.items():
    if status == 'active':
        active_users[user] = status
````

Ejemplo de uso del comando $\mathtt{Break}$

````python
for n in range(2, 10):       # del 2 al 9
    for x in range(2, n):    # del 2 a n.  -> n proviene del loop anterior
        if n % x == 0:
            print(n, 'igual a', x, '*', n//x)
            break
    else: #observar que este else pertenece al for y no al if anterior!!!
        print(n, 'es un número primo')
````

Ejemplo del comando $\mathtt{continue}$

````python
for num in range(2, 10):
    if num % 2 == 0:
        print("Encontró un número par", num)
        continue
    print("Encontró un número impar", num)
````

Qué ocurre si no lo utilizamos

````python
for num in range(2, 10):
    if num % 2 == 0:
        print("Encontró un número par", num)
    print("Encontró un número impar", num)
````
